```python
path = r"G:\AI\AI_train_datasets\FieldBoundaries\PRSS\CodeTests"
os.chdir(path)
```

```python


Sel_metrics = [IOUScore(threshold=0.5), FScore(threshold=0.5)]
opt =  Adam(lr=0.001) #'Adam'
Sel_loss = categorical_focal_dice_loss
print('\n\n ############# Defining and Compiling the model ########### \n')


filepath = r"./sample_models/test_256_4bands_0Over_0Pad.h5"
model = Unet('seresnext101', classes=3, activation="softmax", encoder_weights=None, input_shape=(256, 256, 4))
model.compile(opt, loss=Sel_loss, metrics=Sel_metrics)
print('### Model Compiled ###')

model.load_weights(filepath)
```
